```
import pandas as pd
import numpy as np
import string
```

```
cols = list(range(26)); cols
```

```
# Joining these three vectors into one DataFrame
ucase = [x for x in string.ascii_uppercase]
lcase = [x for x in string.ascii_lowercase]
```

```
alf = pd.DataFrame([lcase,ucase,cols]); alf
```

```
# Transposing this DataFrame (pivoting)
alf = alf.T
```

```
alf.head()
```

```
# Check index
alf.index
```

```
# Changing the column names
alf.columns = ['lowercase','uppercase','number']; alf
```

```
# Selecting some column by name
alf.lowercase
```

```
# Selection some column by slice notation
alf['lowercase']
```

```
# Setting index with range of dates with 26 periods of time
alf.index = pd.date_range('7/1/2020',periods=26); alf
```

```
# Slicing by date
alf['2020-07-12':'2020-07-19']
```

```
# Making another dataframe and printing it
letters = pd.DataFrame({'lowercase':lcase,'uppercase':ucase}); letters.head()
```

```
# Adding another column with randomic values
letters['numbers'] = np.random.randint(1,50,26); letters
```

```
# Check the type of each column
letters.dtypes
```

```
# Setting the values of lowercase column as index
letters.index = lcase; letters
```

```
# Sorting values
letters.sort_values('numbers')
```

```
# Sorting by index
letters.sort_index()
```

```
# Selecting some columns by index notation
letters[['lowercase','uppercase']].head()
```

```
# Selecting by index notation
letters.iloc[5:10]
```
